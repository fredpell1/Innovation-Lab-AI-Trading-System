```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"data1.csv")
df = df.drop("Unnamed: 0.1", axis =1 )
df = df.drop("Unnamed: 0", axis = 1)
```

```python
def williamR(df):
    #fill first 13 positions with 0 since we can't compute a 14 high
    wR = [0 for i in range(13)]
    for i in range(len(df)-13):
        high = max(df["High"][i:i+14])
        low = min(df["Low"][i:i+14])
        wR.append(((high - df["Close"][i+13]) / (high - low))*-1)
    return wR
```

```python
df["WilliamR"] = williamR(df)
```

```python
#Over the whole period
plt.plot(range(1042), df["WilliamR"])
```

```python
#from day 14 to 64 days
plt.plot(range(14,64), df["WilliamR"][14:64])
```

```python
df.to_csv(r"data2.csv")
```
